```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

a = np.array([2,3,5,6])
b = np.arange(0,21,1)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))

a = np.array([2,3,5,6])
b = np.arange(0,6,0.1)

plt.figure()
print("sigmoid:", sigmoid(a))
plt.plot(b, sigmoid(b))

plt.figure()
print("softmax:", softmax(a))
plt.plot(b, softmax(b))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.random.random(100)
y = x * 5 + ((np.random.random(100)-0.5)*0.1) + 20

plt.scatter(x, y)

lr = 0.0001

w, b = np.random.random(2)
print(w, b)

max_iters = 100000
for i in range(100000):
    param_grad = np.sum(((x * w) + b) - y)
    w = w - lr * param_grad
    b = b - lr * (1 if param_grad > 0 else -1)
    
print(w, b)
plt_x = np.array([0, 1])
plt_y = plt_x * w + b
plt.plot(plt_x, plt_y)
```
