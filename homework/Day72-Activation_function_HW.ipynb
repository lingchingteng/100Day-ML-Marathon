# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

```


作業:

    寫出 ReLU & dReLU 一階導數
    並列印


```python

def ReLU(x):
    return abs(x) * (x>0)

x = plt.linspace(-10,10,100)

plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,100)')

plt.grid()

plt.title('ReLU Function')

plt.show()
```

```python

def dReLU(x):
    return 1 * (x>0)

x = plt.linspace(-10,10,100)

plt.plot(x, dReLU(x), 'b', label='linspace(-10,10,100)')

plt.grid()

plt.title('dReLU Function')

plt.show()
```
